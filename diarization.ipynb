```python
from pyannote.audio import Pipeline

# Download and load the pipeline
pipeline = Pipeline.from_pretrained(
    "pyannote/speaker-diarization-community-1",
    token="YOUR_HF_TOKEN_HERE"  # or set HF_TOKEN env var
)

# Optional: use MPS (Apple Silicon GPU)
import torch
if torch.backends.mps.is_available():
    pipeline.to(torch.device("mps"))

# Run diarization
output = pipeline("data/mkbhd_1.mp3")

# Print results
for turn, speaker in output.speaker_diarization:
    print(f"{speaker}: {turn.start:.2f}s - {turn.end:.2f}s")
```

```python
# download the pipeline from Huggingface
from pyannote.audio import Pipeline
pipeline = Pipeline.from_pretrained(
    "pyannote/speaker-diarization-community-1", 
    token="YOUR_HF_TOKEN_HERE")

# run the pipeline locally on your computer
output = pipeline("data/mkbhd_1.mp3")

# print the predicted speaker diarization 
for turn, speaker in output.speaker_diarization:
    print(f"{speaker} speaks between t={turn.start:.3f}s and t={turn.end:.3f}s")
```
